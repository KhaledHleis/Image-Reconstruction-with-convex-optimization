```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

#set the seed for the whole programe
seed = 0
```

```python
original_image = cv.imread("../resources/babun.png",0)
dim = 128
resized_image = cv.resize(original_image,(dim,dim))

plt.imshow(resized_image,'gray')
```

```python
def generate_random_indeces(dim,per,seed = 0):
    """
    this function generates an array of random indeces
    the lenght of the array is determined by a percentage of the image pixel
    the returned shape is (round(dim*dim*per),2)
    """
    np.random.seed(seed)
    return np.floor(np.random.uniform(0,dim,(int (np.round(dim * dim * per)),2))).astype(int)

def mask_Image(image , indeces):
    img = image.copy()
    for i,j in indeces:
        img[i,j] = 0
    return img

dim = resized_image.shape[0]
indeces = generate_random_indeces(dim,0.25)
print(indeces.shape)

Subsample_mask = mask_Image(np.ones((dim,dim)),indeces)
plt.imshow(Subsample_mask,'gray')
plt.show()

subsampled_image = mask_Image(resized_image,indeces)
plt.imshow(subsampled_image,'gray')
plt.show()
```
